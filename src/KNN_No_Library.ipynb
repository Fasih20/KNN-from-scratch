```python
class KNN:
    def __init__(self, k):
        self.k = k

    def fit(self, X, y):
      self.Xtrain = X
      self.ytrain = y

    def distance(self, X1, X2):
      return np.sqrt(np.sum((X1-X2)**2))

    def predict(self, X):
      final_output = []

      for i in range(len(X)):
        d = []
        for j in range(len(self.Xtrain)):
          dist = self.distance(self.Xtrain[j], X[i])
          d.append([dist, j])

        d.sort()
        k_neighbours = d[:self.k]
        votes = [self.ytrain[j] for _, j in k_neighbours]
        ans = max(set(votes), key=votes.count)
        final_output.append(ans)

      return np.array(final_output)
```

```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.neighbors import KNeighborsClassifier #only to compare it with our own model
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Iris.csv')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
feature_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
X =data[feature_columns].values
y =data['Species'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
My_classifier = KNN(k=3)

My_classifier.fit(X_train, y_train)

y_pred1 = My_classifier.predict(X_test)
```

```python
cm1 = confusion_matrix(y_test, y_pred1)
cm1
```

```python
accuracy = accuracy_score(y_test, y_pred1)*100
print('Accuracy of our own coded model is equal to ' + str(round(accuracy, 2)) + ' %.')
```

```python
#Comparing results wit sk-learn's model

classifier = KNeighborsClassifier(n_neighbors=3)

# Fitting the model
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy of Sk-learn model is equal to ' + str(round(accuracy, 2)) + ' %.')
```
